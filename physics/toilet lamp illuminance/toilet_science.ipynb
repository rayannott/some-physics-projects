```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
import warnings
warnings.filterwarnings('ignore')
```

```python
data = np.genfromtxt('toilet_illum.csv',delimiter=',')[6:].T
t, ill = data[0], data[1]
```

```python
def objective(x, a, b):
    return a*np.exp(-b*x)

par, cov = curve_fit(objective, t, ill)
t_fit = np.linspace(0, max(t), 2000)
ill_fit = objective(t_fit, *par)
```

```python
par
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(t, ill)
ax.plot(t_fit, ill_fit, '--')
ax.grid()
plt.xlabel('t, sec')
plt.ylabel('illuminance, lx')
plt.show()
```
